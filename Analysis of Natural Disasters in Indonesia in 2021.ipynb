```python
import pandas as pd

# Membuat dataframe
data = pd.DataFrame({'Banjir': [1435, 732, 342, 455, 476, 380, 171, 328, 59, 61, 109, 1193, 1249, 50, 1176, 419, 39, 187, 570, 864, 735, 929, 343, 200, 288, 567, 620, 240, 259, 125, 153, 270, 138, 204]})

# Menghitung Q1 dan Q3 pada kolom 'A'
Q1 = data['Banjir'].quantile(0.25)
Q3 = data['Banjir'].quantile(0.75)

# Menghitung IQR (Interquartile Range)
IQR = Q3 - Q1

# Menghitung batas atas dan batas bawah
upper = Q3 + (1.5 * IQR)
lower = Q1 - (1.5 * IQR)

# Mendeteksi outlier pada kolom 'A'
outliers = data[(data['Banjir'] < lower) | (data['Banjir'] > upper)]
print(outliers)
```

**Clustering menggunakan algoritma K-Means**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Load data
data = pd.read_csv("bencana alam 2021.csv")
X = data.iloc[:, 1:].values

# Elbow method to determine the number of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()

# K-Means clustering with k = 3
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Visualization
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of Provinces')
plt.xlabel('Number of Disasters')
plt.ylabel('Number of Provinces')
plt.legend()
plt.show()
```

### **PENJELASANNYA DIBAWAH NIH**

Pertama-tama, data dibaca menggunakan pandas dan diambil kolom-kolom dari indeks ke-1 sampai akhir sebagai fitur-fitur yang akan digunakan dalam clustering. Kemudian, dilakukan elbow method untuk menentukan jumlah kluster yang optimal. Elbow method dilakukan dengan memvariasikan jumlah kluster dari 1 hingga 10 dan menghitung within-cluster sum of squares (WCSS) untuk setiap nilai k. Hasilnya direpresentasikan dalam sebuah plot, dan jumlah kluster yang optimal adalah titik siku pada kurva. Dalam contoh ini, titik siku berada pada k = 3, sehingga nilai k = 3 dipilih untuk clustering.

Setelah nilai k ditentukan, algoritma K-Means dijalankan dengan jumlah kluster yang telah ditentukan. Setiap baris data diberi label kluster 0, 1, atau 2 berdasarkan kluster mana yang memiliki centroid terdekat dengan baris tersebut.

Hasil clustering divisualisasikan dalam sebuah scatter plot, di mana setiap titik merepresentasikan sebuah provinsi. Setiap kluster ditandai dengan warna yang berbeda, dan pusat kluster (centroid) ditandai dengan lingkaran kuning besar. Pada plot ini, sumbu x merepresentasikan jumlah bencana alam yang terjadi, sedangkan sumbu y merepresentasikan jumlah provinsi.

```python
import pandas as pd

#membaca dataset
data = pd.read_csv("bencana alam 2021.csv")

# Menghitung kuartil dan IQR
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas bawah dan batas atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Mengidentifikasi outlier
outliers = data[(data < lower_bound) | (data > upper_bound)].stack()

# Menampilkan outlier
print("Outlier:")
print(outliers)
```

### **PENJELASANNYA DIBAWAH NIH**
Dari output tersebut dapat dilihat bahwa terdapat beberapa data yang dianggap sebagai outlier, yaitu data dengan nilai di luar batas bawah dan batas atas yang ditentukan dengan menggunakan metode IQR.

```python
import pandas as pd
import numpy as np

#membaca dataset
data = pd.read_csv("bencana alam 2021.csv")

# hitung IQR untuk setiap kolom
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

# tentukan batas atas dan bawah untuk outlier
upper_limit = Q3 + 1.5*IQR
lower_limit = Q1 - 1.5*IQR

# cari data yang di luar batas atas dan bawah
outliers = ((data < lower_limit) | (data > upper_limit)).any(axis=1)

# tampilkan data yang di luar batas atas dan bawah
print(data[outliers])

# ganti nilai outlier dengan nilai yang lebih realistis (misalnya median)
data[outliers] = np.nanmedian(data)
```

### _**PEMODELAN DATA**_

```python
#membaca dataset
data = pd.read_csv("bencana alam 2021.csv")
data
```

```python
#dikarenakan variabel 'Provinsi' tipe datanya string, maka variabel tersebut dihapus
#untuk melakukan pemodelan data
# menghapus variabel "Provinsi"
data = data.drop('Provinsi', axis=1)

# menampilkan dataset setelah variabel dihapus
print(data)
```

```python
#transforming variabel prediktor dengan metode MinMaxScaler
from sklearn import preprocessing
tranformasi_min_max = preprocessing.MinMaxScaler()
```

```python
#membagi data menjadi variabel target dan atribut
target = 'Banjir'

x = data.drop('Banjir', axis=1)
y = data[target]
nama_var_x = ['Tanah Longsor', 'Banjir Bandang', 'Gempa Bumi', 'Tsunami', 'Gelombang Pasang Laut', 'Angin Puyuh / Angin Puting Beliung / Topan', 'Gunung Meletus', 'Kebakaran Hutan', 'Kekeringan', 'Tidak Ada Bencana Alam']
x = pd.DataFrame(data=tranformasi_min_max.fit_transform(x), columns=nama_var_x)
```

```python
#membagi data menjadi data training dan data testing (20% training dan 80% testing)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.80,random_state=1)
```

```python
#memodelkan data dengan regresi linier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lin_model = LinearRegression()
lin_model.fit(x_train, y_train)
```

```python
#tranformasi variabel menjadi polinomial
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
polynomial_features = PolynomialFeatures(degree=2)
x_poly = polynomial_features.fit_transform(x_train)
poly_model = LinearRegression()
poly_model.fit(x_poly, y_train)
x_poly_test = polynomial_features.fit_transform(x_test)
```

```python
from sklearn.linear_model import Ridge

#model ridge dengan alpha = 1
model_ridge = Ridge(alpha = 1)
model_ridge.fit(x_train, y_train)
```

```python
from sklearn.linear_model import Lasso

#model lasso dengan alpha=1
model_lasso = Lasso(alpha=1)
model_lasso.fit(x_train, y_train)
```

### _**DATA TRAINING DAN DATA TESTING REGRESI LINEAR**_

```python
#evaluasi model untuk data training
y_train_predict = lin_model.predict(x_train)
rmse = (np.sqrt(mean_squared_error(y_train, y_train_predict)))
r2 = r2_score(y_train, y_train_predict)

print("Performa Model Data Training")
print("==============================")
print("RMSE : {}".format(rmse))
print("R2 : {}".format(r2))
```

```python
#evaluasi model untuk data testing
y_test_predict = lin_model.predict(x_test)
rmse_1 = (np.sqrt(mean_squared_error(y_test, y_test_predict)))
r2_1 = r2_score(y_test, y_test_predict)

print("Performa Model Data Testing")
print("==============================")
print("RMSE : {}".format(rmse_1))
print("R2 : {}".format(r2_1))
```

### _**Data Training & Testing Regresi Polinomial**_

```python
#evaluasi model untuk data training
y_train_pred_poly = poly_model.predict(x_poly)
rmse = (np.sqrt(mean_squared_error(y_train, y_train_pred_poly)))
r2 = r2_score(y_train, y_train_pred_poly)

print("Performa Model Polinomial Derajat 2 untuk Data Training")
print("==========================================================")
print("RMSE : {}".format(rmse))
print("R2 : {}".format(r2))
```

```python
#evaluasi model untuk data testing
y_test_pred_poly = poly_model.predict(x_poly_test)
rmse_1 = (np.sqrt(mean_squared_error(y_test, y_test_pred_poly)))
r2_1 = r2_score(y_test, y_test_pred_poly)

print("Performa Model Polinomial Derajat 2 untuk Data Testing")
print("==========================================================")
print("RMSE : {}".format(rmse_1))
print("R2 : {}".format(r2_1))
```

### _**Data Training & Testing Regresi Ridge**_

```python
#evaluasi model untuk data training
y_train_pred_ridge = model_ridge.predict(x_train)
rmse = (np.sqrt(mean_squared_error(y_train, y_train_pred_ridge)))
r2 = r2_score(y_train, y_train_pred_ridge)

print("Performa Model Ridge untuk Data Training")
print("============================================")
print("RMSE : {}".format(rmse))
print("R2 : {}".format(r2))
```

```python
#evaluasi model untuk data testing
y_test_pred_ridge = model_ridge.predict(x_test)
rmse_1 = (np.sqrt(mean_squared_error(y_test, y_test_pred_ridge)))
r2_1 = r2_score(y_test, y_test_pred_ridge)

print("Performa Model Ridge untuk Data Testing")
print("============================================")
print("RMSE : {}".format(rmse_1))
print("R2 : {}".format(r2_1))
```
